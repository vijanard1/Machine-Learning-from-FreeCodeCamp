```
#using a different model from tensor flow this time, 
# deals with probabilties
import tensorflow_probability as tfp 
import tensorflow as tf
```

1. Cold day is encoded as 0 and hot day is encded as 1
2. First day is 80% chance of being cold day
3. Cold day is followed by hot day with probability of 30%
Hot day has 20% chance of being followed by cold day
4. On each day the temperature is normally distributed 
with a mean and standard deviation of 0 and 5 for cold day, min = -5 max = 5
with a mean and standard deviation of 15 and 10 for cold day, min = 5 , max= 25



```
tfd = tfp.distributions
# initial_distribution = tfd.Categorical(probs = [0.8, 0.2]) #point 2, first day being 80% cold day
initial_distribution = tfd.Categorical(probs = [0.2, 0.8]) #point 2, first day being 20% cold day, starting with a hot day

transition_distribution = tfd.Categorical(probs=[[0.7,0.3], #50% chance of hot day following a cold day
# transition_distribution = tfd.Categorical(probs=[[0.7,0.3], 
                                                 
                                                 [0.2,0.8]]) # point 3
observation_distribution = tfd.Normal(loc=[0.,15], scale=[5.,10.])  # loc = mean values and scale =stand deviation values
```

```
model = tfd.HiddenMarkovModel(
    initial_distribution=initial_distribution,
    transition_distribution=transition_distribution,
    observation_distribution=observation_distribution,
    num_steps=7)
# num_steps = no of steps we need to run the model ,here it is the no of days we want to predcit the weather for 
```

```
mean = model.mean()
# we are calculating the average temperature for each day

# due to the way TF operates on a lower level we need to evaluate part of the graph, ie tensor
# from with in a session
with tf.compat.v1.Session() as sess:
  print(mean.numpy())
```
